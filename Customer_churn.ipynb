### DATA PREPROCESSING and FEATURE ENGINEERING

```python
# To change the values in HasCrCard, IsActiveMember from integer to object datatype
# Create a function to replace 1 with 'yes' and 0 with 'no'

def map_values(column):
    return column.map({1: 'Yes', 0: 'No'})

# Applying the mapping function to 'IsActiveMember' and 'HasCrCard' columns
df['HasCrCard'] = map_values(df['HasCrCard'])
df['IsActiveMember'] = map_values(df['IsActiveMember'])

# Check updated dataframe
print(df)
```

```python
# Visalize altered column 
print(df[['HasCrCard', 'IsActiveMember']])
```

```python
# To change the values in Exited columns from a numeric to object datatype
 
# Create a function to map 1 to 'Churned' and 0 to 'Retained'
def map_values_exited(column):
    return column.map({1: 'Churned', 0: 'Retained'})

# Applying the mapping function to 'Exited columns
df['Exited'] = map_values_exited(df['Exited'])

# Check updated dataframe
print(df)

```

```python
# Creating a new column called 'age_group' to aggregate customer ages into specified categories.
def categorize_age(age):
    if 18 <= age <= 35:
        return 'Young Adults(18-35)' # For ages 18 - 35
    elif 36 <= age <= 60:
        return 'Middle Aged(36-60)' # For ages 36 - 60
    elif age >= 61:
        return 'Seniors(61 and above)' # For ages above 60
    else:
        return 'Out of Range'  # For ages less than 18 

# Apply the function to the "age" column to create a new "age_group" column
df['age_group'] = df['Age'].apply(categorize_age)

# Display results
print(df)
```

```python
# Age distribution for all customers

# Get the value counts for the categorized age groups
age_group_counts = df['age_group'].value_counts().reset_index()

# Rename the columns for clarity
age_group_counts.columns = ['Age Group', 'Count']

# Add proportions for each age group
age_group_counts['Proportion (%)'] = (age_group_counts['Count'] / df.shape[0]) * 100

# Display the results
print("Categorized Age distribution for all customers:")
print(age_group_counts)
```

### Initial tentative observation

The demographic of customers were categorized into three main age groups highlighted by different stages in adulthood

`Young Adults (18-35)`: Represents the youngest segment of the customer base.

`Middle-Aged (36-60)`: Covers the majority of working-age adults.

`Seniors (Above 60)`: Represents the older demographic segment.

Middle Aged Adults accounted for more than half of total customer base at `53.8%` while Young Adults followed closely at `41.5%`; the Seniors were significamtly smaller in size at just `4.7%`. 

This clearly suggests that Middle Aged Adults forms the core demographic for the company showing that the companyâ€™s products/services may appeal most strongly to this group.

The significant presence of middle-aged customers underscores the importance of maintaining loyalty in this group, while the smaller senior segment highlights an opportunity to grow in an underserved market.

```python
# Isolating the customer age group with the highest churn 

# Group by age_group and Exited, then calculate the churn count
churn_counts = df.groupby(['age_group', 'Exited']).size().unstack(fill_value=0)

# Add total customers column
churn_counts['Total_Customers'] = churn_counts['Churned'] + churn_counts['Retained']

# Calculate churn and retention rates for each age group
churn_counts['Churn_Rate'] = (churn_counts['Churned'] / churn_counts['Total_Customers'] * 100).round(2)
churn_counts['Retention_Rate'] = (churn_counts['Retained'] / churn_counts['Total_Customers'] * 100).round(2)

# Calculate percentage of total customers in each age group
total_customers = churn_counts['Total_Customers'].sum()
churn_counts['Percent_of_Total'] = (churn_counts['Total_Customers'] / total_customers * 100).round(2)

# Identify the age groups with highest churn and retention rates
highest_churn_group = churn_counts['Churn_Rate'].idxmax()
highest_retention_group = churn_counts['Retention_Rate'].idxmax()

# Display the results
print("Customer Age Group vs churn Analysis:")
print("\nDetailed Counts and Rates:")
print("-------------------------")
print(churn_counts[['Churned', 'Retained', 'Total_Customers', 'Percent_of_Total', 'Churn_Rate', 'Retention_Rate']])
print("\nSummary:")
print(f"Total Customers: {total_customers}")
print(f"Age group with highest churn rate: {highest_churn_group} ({churn_counts.loc[highest_churn_group, 'Churn_Rate']}%)")
print(f"Age group with highest retention rate: {highest_retention_group} ({churn_counts.loc[highest_retention_group, 'Retention_Rate']}%)")
```

### Initial tentative observation

Customers were categorized based on their age groups i.e Young Adults, Middle Aged and Senior Adults.

Middle Aged Adults had the highest churn rate at `29%` nearly three times that of the young adults at `8%` but just slightly above the Seniors at `25%` which points to the fact that customers in this category might be more demanding and financially affluent, leading to dissatisfaction and eventual churn if services don't meet expectations. 

It is worth noting that Seniors still had a relatively high churn rate just slightly below Middle Aged adults; most likey due to their assumed financial limitations, retirement, or shifting preferences, further analysis would be required to confirm this.

Young Adults were the most loyal by a distance `92% retained`; They represent an opportunity for growth through targeted marketing and retention strategies.





```python
# Categorizing Salaries of all customers to discover the distribution of their earnings

# Function to categorize EstimatedSalary
def categorize_salary(salary):
    if 11.580 <= salary <= 66671.88:
        return 'Low salary'  # $11.58 - $66,671.88
    elif 66671.88 < salary <= 133332.18:
        return 'Medium salary'  # $66,671.88 - $133,332.18
    elif 133332.18 < salary <= 199992.48:
        return 'High salary'  # $133,332.18 - $199,992.48
    else:
        return 'Out of Range'  # Just in case there are any unexpected values

# Apply the function to create a new column
df['salary_category'] = df['EstimatedSalary'].apply(categorize_salary)

# Display the first few rows of the dataset to confirm the new column
print("First few rows with the salary category:")
print(df[['EstimatedSalary', 'salary_category']].head())

# Count the number of customers in each category
salary_category_counts = df['salary_category'].value_counts()

# Convert counts to a DataFrame and calculate proportions
salary_category_df = salary_category_counts.to_frame(name='Count')
salary_category_df['Proportion (%)'] = (salary_category_df['Count'] / df.shape[0]) * 100

print("\nDistribution of salary categories with proportions:")
print(salary_category_df)
```

### Initial tentative observation

The Estimated salary of all customers was categorized based on level of their earnings:

`Low Salary (between $11.58 - $66,671.88)`: Customers in this category are likely in entry-level jobs, part-time employment, or lower-paying industries.

`Medium Salary ($66,671.88 - $133,332.18)`: Customers that fall into this category, are more likely financially stable and composed of middle-aged individuals in mid-level positions or professionals with consistent earnings. They are usually able to access a broader range of financial products, such as savings plans, loans, or moderate investment options.

`High Salary ($133,332.18 - $199,992.48)`: Customers in this bracket make the highest earnings, potentially senior professionals, business owners, or retirees with significant financial resources. They are likely to seek premium financial services or luxury-focused services.


There are very slight differences between the number of customers in these categories with those in the Medium class being the most predominant at `3383 (33.83% of the customer base)` followed closely by Low salary class at `3314 (33.14% of the customer base)`, with the Highest earners being the fewest at a little over `3000 customers (33.03%)`.

There is an even distribution amongst these income groups meaning all are well represented overall. 
The "Medium" salary category has the highest representation, indicating that most customers fall within this income bracket.

```python
# Age groups vs. salary categories analysis

# Grouping data by age groups and salary categories
age_salary_analysis = df.groupby(['age_group', 'salary_category']).size().unstack(fill_value=0)

# Adding a total count per age group
age_salary_analysis['Total'] = age_salary_analysis.sum(axis=1)

# Calculating proportions for each salary category within each age group
age_salary_analysis['Low salary (%)'] = (age_salary_analysis['Low salary'] / age_salary_analysis['Total']) * 100
age_salary_analysis['Medium salary (%)'] = (age_salary_analysis['Medium salary'] / age_salary_analysis['Total']) * 100
age_salary_analysis['High salary (%)'] = (age_salary_analysis['High salary'] / age_salary_analysis['Total']) * 100

# Identifying the age group with the highest count in each salary category
max_low_group = age_salary_analysis['Low salary'].idxmax()
max_medium_group = age_salary_analysis['Medium salary'].idxmax()
max_high_group = age_salary_analysis['High salary'].idxmax()

# Display the detailed table
print("Age groups vs. salary categories analysis:")
print(age_salary_analysis)

print("\nObservations:")
print(f"The age group with the most low salary earners is: {max_low_group}")
print(f"The age group with the most medium salary earners is: {max_medium_group}")
print(f"The age group with the most high salary earners is: {max_high_group}")
```

`Middle Aged customers` dominated in all earning categories ranking the highest overall. Having a substantial number of customers in the `Medium salary category at 35%`, makes it the age group with the highest count of Mid-income earners.
It also maintains a healthy proportion in the `High salarycategory (33%)` while having the lowest percentage in the `Low salary category (32%)`. Surprisingly `Young adults` came a close second in all categories which on deeper analysis can be attributed to their large numbers compared to seniors; they comprise a significant portion of the dataset and were predominantly but unsurprsingly found in the `Low salary category (34%)`; being younger individuals, they are often early in their careers and are still building their earning potential.

```python
# Customer Salary Category vs. Churn Analysis

# Group by salary_category and Exited, then calculate the churn count
salary_churn_counts = df.groupby(['salary_category', 'Exited']).size().unstack(fill_value=0)

# Add total customers column
salary_churn_counts['Total_Customers'] = salary_churn_counts['Churned'] + salary_churn_counts['Retained']

# Calculate churn and retention rates for each salary category
salary_churn_counts['Churn_Rate'] = (salary_churn_counts['Churned'] / salary_churn_counts['Total_Customers'] * 100).round(2)
salary_churn_counts['Retention_Rate'] = (salary_churn_counts['Retained'] / salary_churn_counts['Total_Customers'] * 100).round(2)

# Calculate percentage of total customers in each salary category
total_customers = salary_churn_counts['Total_Customers'].sum()
salary_churn_counts['Percent_of_Total'] = (salary_churn_counts['Total_Customers'] / total_customers * 100).round(2)

# Identify the salary categories with highest churn and retention rates
highest_churn_salary = salary_churn_counts['Churn_Rate'].idxmax()
highest_retention_salary = salary_churn_counts['Retention_Rate'].idxmax()

# Display the results
print("Customer Salary Category vs. Churn Analysis:")
print("\nDetailed Counts and Rates:")
print("-------------------------")
print(salary_churn_counts[['Churned', 'Retained', 'Total_Customers', 'Percent_of_Total', 'Churn_Rate', 'Retention_Rate']])
print("\nSummary:")
print(f"Total Customers: {total_customers}")
print(f"Salary category with highest churn rate: {highest_churn_salary} ({salary_churn_counts.loc[highest_churn_salary, 'Churn_Rate']}%)")
print(f"Salary category with highest retention rate: {highest_retention_salary} ({salary_churn_counts.loc[highest_retention_salary, 'Retention_Rate']}%)")
```

```python
# Customer Gender vs. Churn Analysis 

# Grouping data by gender and Exited status
gender_churn_analysis = df.groupby(['Gender', 'Exited']).size().unstack(fill_value=0)

# Adding total customers column
gender_churn_analysis['Total_Customers'] = gender_churn_analysis['Churned'] + gender_churn_analysis['Retained']

# Calculate churn and retention rates for each gender
gender_churn_analysis['Churn_Rate'] = (gender_churn_analysis['Churned'] / gender_churn_analysis['Total_Customers'] * 100).round(2)
gender_churn_analysis['Retention_Rate'] = (gender_churn_analysis['Retained'] / gender_churn_analysis['Total_Customers'] * 100).round(2)

# Calculate percentage of total customers in each gender
total_customers = gender_churn_analysis['Total_Customers'].sum()
gender_churn_analysis['Percent_of_Total'] = (gender_churn_analysis['Total_Customers'] / total_customers * 100).round(2)

# Identify the gender with highest churn and retention rates
highest_churn_gender = gender_churn_analysis['Churn_Rate'].idxmax()
highest_retention_gender = gender_churn_analysis['Retention_Rate'].idxmax()

# Display the results
print("Customer Gender vs. Churn Analysis:")
print("\nDetailed Counts and Rates:")
print("-------------------------")
print(gender_churn_analysis[['Churned', 'Retained', 'Total_Customers', 'Percent_of_Total', 'Churn_Rate', 'Retention_Rate']])
print("\nSummary:")
print(f"Total Customers: {total_customers}")
print(f"Gender with highest churn rate: {highest_churn_gender} ({gender_churn_analysis.loc[highest_churn_gender, 'Churn_Rate']}%)")
print(f"Gender with highest retention rate: {highest_retention_gender} ({gender_churn_analysis.loc[highest_retention_gender, 'Retention_Rate']}%)")

```

```python
# Customer Gender vs. Salary Category Analysis

# Grouping data by gender and salary categories
gender_salary_analysis = df.groupby(['Gender', 'salary_category']).size().unstack(fill_value=0)

# Adding total customers column
gender_salary_analysis['Total_Customers'] = gender_salary_analysis.sum(axis=1)

# Calculate percentage distribution of salary categories within each gender
for category in ['Low salary', 'Medium salary', 'High salary']:
    gender_salary_analysis[f'{category} (%)'] = (gender_salary_analysis[category] / gender_salary_analysis['Total_Customers'] * 100).round(2)

# Calculate percentage of total customers by gender
total_customers = gender_salary_analysis['Total_Customers'].sum()
gender_salary_analysis['Percent_of_Total'] = (gender_salary_analysis['Total_Customers'] / total_customers * 100).round(2)

# Identify the predominant salary category for each gender
predominant_categories = gender_salary_analysis[['Low salary', 'Medium salary', 'High salary']].idxmax(axis=1)

# Display the results
print("Customer Gender vs. Salary Category Analysis:")
print("\nDetailed Counts and Percentages:")
print("-------------------------")
print(gender_salary_analysis[['Low salary', 'Medium salary', 'High salary', 
                             'Total_Customers', 'Percent_of_Total',
                             'Low salary (%)', 'Medium salary (%)', 'High salary (%)']])
print("\nSummary:")
print(f"Total Customers: {total_customers}")
print("\nPredominant salary category by gender:")
for gender in predominant_categories.index:
    print(f"{gender}: {predominant_categories[gender]} "
          f"({gender_salary_analysis.loc[gender, predominant_categories[gender]]} customers, "
          f"{gender_salary_analysis.loc[gender, f'{predominant_categories[gender]} (%)']}%)")

```
